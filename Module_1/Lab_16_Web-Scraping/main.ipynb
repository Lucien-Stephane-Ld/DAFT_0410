#### Display the names of the trending developers retrieved in the previous step.

Your output should be a Python list of developer names. Each name should not contain any html tag.

**Instructions:**

1. Find out the html tag and class names used for the developer names. You can achieve this using Chrome DevTools.

1. Use BeautifulSoup to extract all the html elements that contain the developer names.

1. Use string manipulation techniques to replace whitespaces and linebreaks (i.e. `\n`) in the *text* of each html element. Use a list to store the clean names.

1. Print the list of names.


````python
url = 'https://github.com/trending/developers'
response = rq.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

dev_tags = soup.find_all('h1', {'class': 'h3 lh-condensed'})

dev_repos_tags = soup.find_all('h1', {'class': 'h4 lh-condensed'})

for dev_tag, repo_tag in zip(dev_tags, dev_repos_tags):
    dev_name = dev_tag.text.strip()
    repo_name = repo_tag.text.strip()
    print(f"{dev_name} ({repo_name})")
````

````python
url = 'https://github.com/trending/developers'
response = rq.get(url)
soup = BeautifulSoup(response.content, 'html.parser')


name_filt = 'div:nth-child(1) > h1 > a'
dev_name = []
for element in soup.select(name_filt):
    dev_name.append(element.get_text().strip('\n').strip(' '))

dev_name
````

````python
tag_filt = 'div:nth-child(1) > p > a'
tag_name = []
for element in soup.select(tag_filt):
    tag_name.append(element.get_text().strip('\n').strip(' '))
````

#### Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

repo_select = 'h2 > a'
repo_lst = []
for element in soup.select(repo_select):
    repo_lst.append(element.get("href").strip("/"))

repo_df = pd.DataFrame({'Repository Path' : repo_lst})
repo_df
````

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
````

````python
url = 'https://www.imdb.com/chart/top'

response = requests.get(url).content

imdb_soup = BeautifulSoup(response, "html.parser")
print(imdb_soup.prettify())
````

````python
selector = '#main > div > span > div > div > div.lister > table > tbody > tr:nth-child(1) > td.titleColumn > a'

movie_title = imdb_soup.select(selector)#.get_text()  removes everything in between <>

print(movie_title)
````

````python
selector2 = 'td.titleColumn > a'

movie_lst = []

for link in imdb_soup.select(selector2):
    movie_lst.append(link.get_text())

year_selector = 'td.titleColumn > span'

year_lst = []

for link in imdb_soup.select(year_selector):
    i = link.get_text().strip(")(")
    year_lst.append(i)
````

````python
movie_df = pd.DataFrame({'movie_title' : movie_lst, 'release_year' : year_lst})

movie_df.columns = ['Title','Year']
````
